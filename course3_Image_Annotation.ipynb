```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread('images/natur.jpg', cv.IMREAD_COLOR)
plt.imshow(img[:,:,::-1])
```

Drawing a line

```python
imgLine = img.copy()
cv.line(imgLine,(50,120),(100,120),(0,0,255),thickness=1,lineType=cv.LINE_8)
plt.imshow(imgLine[:,:,::-1])
```

Drawing a circle

```python
imgCirle = img.copy()
cv.circle(imgCirle,(175,100),radius=20,color=(0,255,0),thickness=2,lineType=cv.LINE_AA)
plt.imshow(imgCirle[:,:,::-1])
```

Drawing a rectangle

```python
imgRect = img.copy()
cv.rectangle(imgRect,pt1=(140,60),pt2=(200,140),color=(255,0,0),thickness=2,lineType=cv.LINE_AA)
plt.imshow(imgRect[:,:,::-1])
```

Putting text on the image

```python
from networkx import center


imgText = img.copy()
text= "image annotation with opencv"
FontScale = 0.5
FontFace = cv.FONT_HERSHEY_SIMPLEX
FontColor = (0,255,255)
cv.putText(imgText,text,(40,20),FontFace,FontScale,FontColor,thickness=1,lineType=cv.LINE_AA)
plt.imshow(imgText[:,:,::-1])
```
