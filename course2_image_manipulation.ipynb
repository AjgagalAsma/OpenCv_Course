```python
import cv2
import matplotlib.pyplot as plt

# Read image as gray scale.
cb_img = cv2.imread("images/img.jpg", 0)

# Set color map to gray scale for proper rendering.
plt.imshow(cb_img, cmap="gray")
print(cb_img)
```

Accessing Individual Pixels

```python
# print the first pixel of the first black box
print(cb_img[0, 0])
# print the first white pixel to the right of the first black box
print(cb_img[0, 40])
```

Modifying Image Pixels

```python
cb_img_copy = cb_img.copy()

cb_img_copy[10, 10] = 255
cb_img_copy[10, 40] = 0 
plt.imshow(cb_img_copy, cmap="gray")
```

Cropping Images

```python
img = cv2.imread("images/natur.jpg", 1)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
crop_img = img_rgb[40:120,80:200]
plt.imshow(crop_img)
```

Resizing Images

```python
#method 1
new_crop_img = cv2.resize(crop_img,None,fx=2,fy=2)
plt.imshow(new_crop_img)
print(new_crop_img.shape)
```

```python
#method 2
new_crop_img = cv2.resize(crop_img, dsize=(240, 160), interpolation=cv2.INTER_AREA)
plt.imshow(new_crop_img)
```

```python
#change the imge size while keeping the aspect ratio
plt.imshow(crop_img)
```

```python
dezired_width = 200
aspect_ratio = dezired_width / crop_img.shape[1]
new_height = int(crop_img.shape[0]*aspect_ratio)
new_crop_img = cv2.resize(crop_img, dsize=(dezired_width,new_height), interpolation=cv2.INTER_CUBIC)
plt.imshow(new_crop_img)
```

Flipping Images

```python
img_rgb_flipped_horz = cv2.flip(img_rgb, 1)
img_rgb_flipped_vert = cv2.flip(img_rgb, 0)
img_rgb_flipped_both = cv2.flip(img_rgb, -1)

# Show the images
plt.figure(figsize=(18, 5))
plt.subplot(141);plt.imshow(img_rgb_flipped_horz);plt.title("Horizontal Flip");
plt.subplot(142);plt.imshow(img_rgb_flipped_vert);plt.title("Vertical Flip");
plt.subplot(143);plt.imshow(img_rgb_flipped_both);plt.title("Both Flipped");
plt.subplot(144);plt.imshow(img_rgb);plt.title("Original");
```
